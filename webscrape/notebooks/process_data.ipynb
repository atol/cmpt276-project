```python
import pandas as pd
from bs4 import BeautifulSoup, Comment
```

```python
df = pd.read_csv('../data/travel.csv')
```

```python
df
```

```python
# Extract relevant columns
df1 = df[['country', 'advisory', 'entryexit', 'link', 'updated']].copy()
```

```python
df1
```

```python
def find_passport(text):
    soup = BeautifulSoup(text, 'lxml')
    pass_info = ""
    
    try:
        target = soup.find('h4', text='Regular Canadian passport')
        for sibling in target.find_next_siblings():
            pass_info += sibling.text
            if sibling.name != (None, 'p'):
                break;
    except:
        pass
    
    return pass_info
```

```python
# Extract passport requirements
passport_info = df1['entryexit'].apply(lambda text : find_passport(text))
```

```python
df1['passport'] = passport_info
```

```python
df1
```

```python
def find_visas(text):
    soup = BeautifulSoup(text, 'lxml')
    visa_info = []
    
    try:
        target = soup.find('h3', text='Visas')
        for sibling in target.find_next_siblings():
            visa_info.append(sibling)
            if sibling.name != (None, 'p'):
                break;
    except:
        pass
    
    return visa_info
```

```python
# Extract visa requirements
visa_info = df1['entryexit'].apply(lambda text : find_visas(text))
```

```python
df1['visas'] = visa_info
```

```python
df1
```

```python
df1['visas'].iloc[0]
```

```python
# Convert list to string
df1['visas'] = df1['visas'].apply(lambda x : ','.join(map(str, x)))
```

```python
df1
```

```python
def replace_line(text):
    soup = BeautifulSoup(text, 'lxml')
    for br in soup.find_all("br"):
        br.replace_with("\n")
    return soup
```

```python
# Replace <br> tags with new lines
df1['visas'] = df1['visas'].apply(lambda text : replace_line(text))
```

```python
df1
```

```python
df1['visas'].iloc[0]
```

```python
# Convert list to string
df1['visas'] = df1['visas'].apply(lambda x : ','.join(map(str, x)))
```

```python
df1
```

```python
# Remove HTML tags
df1.visas = df1.visas.str.replace('<[^<]+?>', '')
```

```python
df1
```

```python
# Get final columns
df2 = df1[['country', 'advisory', 'passport', 'visas', 'link', 'updated']].copy()
```

```python
df2
```

```python
# Sort alphabetically by country
df3 = df2.sort_values('country')
```

```python
# Save to .csv file
df3.to_csv("../data/travel_clean.csv", index=False)
```
